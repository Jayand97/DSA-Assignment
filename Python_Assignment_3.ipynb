```python
# Line Plot Representaion
data.plot()
plt.title("Line Plot")
plt.show()
```

```python
# Scatter plot with Classification and SL
data.plot(kind = 'scatter',x ='Classification', y = 'SL')
plt.title(" Classification  and SL Scatter plot")
plt.show()
```

```python
# Histogram Plot
data.plot(kind ='hist',width=0.4)

plt.title("Histogram Plot")
plt.show()
```

```python
# Bar plot for PL and Classification
plt.bar(data['PL'],data['Classification'],width=0.5 )

plt.xticks(rotation = 45)
plt.title("Barchart Plot")
plt.show()
```

```python
# Box plot for SW
plt.boxplot(data['SW'])
plt.show()

```
