Outlier in Age Column


```python
# Checking outlier before imputing
sns.boxplot(x = data['Age'])
```

```python
# Teating Outlier by Imputing

q1= data['Age'].quantile(0.25)
q3 = data['Age'].quantile(0.75)

iqr = q3- q1
iqr
```

```python
upper_limit = q3 + 1.5*iqr
lower_limit = q1 -1.5*iqr

print(upper_limit)
print(lower_limit)
```

```python
def limit_imputer(value):
  if value>upper_limit:
    return upper_limit
  if value<lower_limit:
    return lower_limit
  else:
    return value
```

```python
# applying the imputer function
data['Age'] = data['Age'].apply(limit_imputer)
```

```python
data['Age'].describe()
```

```python
# Checking outlier after imputing
sns.boxplot(x = data['Age'])
```
