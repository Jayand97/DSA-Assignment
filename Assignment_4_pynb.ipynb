# **Case Study on Testing on Hypothesis Using Sales_add.csv**

```python
from scipy.stats import ttest_rel, chi2_contingency
```

```python
data= pd.read_csv('/content/Sales_add.csv')
data
```

```python
data2 = data.rename(columns={'Sales_before_digital_add(in $)':'Sales_Before_DM','Sales_After_digital_add(in $)':'Sales_After_DM'})
data2
```

```python
# Increase in Sales After Digital Marketing

# Perform a paired T-test to compare sales before and after digital marketing
sales_before = data2['Sales_Before_DM']
sales_after = data2['Sales_After_DM']
t_stat, p_value = ttest_rel(sales_before, sales_after)
p_value
```

```python
print(f"Paired t-test results:\n t-statistic: {t_stat}\n p-value: {p_value}")
if p_value < 0.05:
    print("There is a significant increase in sales after stepping into digital marketing.")
else:
    print("There is no significant increase in sales after stepping into digital marketing.")
```

```python
# Dependency Between 'Region' and 'Manager'
# Create a contingency table and perform a Chi-squared test
tab_data = pd.crosstab(data2['Region'], data2['Manager'])
chi2stat, p_value, dof, exp = chi2_contingency(tab_data)
p_value
```

```python
print(f"\nChi-squared test results:\nChi2-statistic: {chi2stat}\np-value: {p_value}")
if p_value < 0.05:
    print("There is a dependency between the features 'Region' and 'Manager'.")
else:
    print("There is no dependency between the features 'Region' and 'Manager'.")
```
